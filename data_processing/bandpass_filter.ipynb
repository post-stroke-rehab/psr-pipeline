```python
import numpy as np
import matplotlib.pyplot as plt
from bandpass_filter import bandpass_filter   

np.random.seed(0)

def rms(x):
    return np.sqrt(np.mean(x**2))

fs = 2000
duration = 1.0
t = np.arange(0, duration, 1/fs)

#my synthtetic signal: 10 Hz (out of band), 100 & 300 Hz (in band), & noise
sig = (0.5*np.sin(2*np.pi*10*t) +
       1.0*np.sin(2*np.pi*100*t) +
       0.7*np.sin(2*np.pi*300*t) +
       np.random.normal(0, 0.05, len(t)))

filtered = bandpass_filter(sig, fs=fs, low=20, high=450, order=4, zero_phase=True)

print("Raw RMS:", rms(sig))
print("Filtered RMS:", rms(filtered))

plt.figure(figsize=(10, 4))
plt.plot(t[:2000], sig[:2000], label="raw", alpha=0.5)
plt.plot(t[:2000], filtered[:2000], label="filtered")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Bandpass filter test")
plt.legend()
plt.tight_layout()
plt.show()
```
